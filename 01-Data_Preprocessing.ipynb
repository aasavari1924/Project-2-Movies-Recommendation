```python
# importing Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# importing data set

movies = pd.read_csv("movies.csv")
credits = pd.read_csv("credits.csv")
```

```python
# join 2 data set with same column name

merge_df = pd.merge(movies, credits, left_on = 'id', right_on = 'movie_id', how = 'inner')
```

```python
new_df = merge_df[['id','genres','keywords','overview','title_x','cast','crew','production_companies']]
new_df.head(2)
```

```python
# changing columns heading
new_df = new_df.rename(columns = {'title_x' : 'title'})
```

```python
new_df.loc[0, 'keywords'] 
```

```python
# using function convert string to comma seperated values

import ast
def get_names(obj):
    x = ast.litteral_eval(obj)
    keywords = []
    for r in x:
        name = r["name"].replace(" ","_"),lower().strip()
        if name not in keywords:
            keywords.append(name)
    return ",".join(keywords)
```

```python
new_df["keywords"] = new_df["keywords"].apply(get_names)
```

```python
new_df["genres"] = new_df["genres"].apply(get_names)
```

```python
new_df["cast"] = new_df["cast"].apply(get_names)
```

```python
new_df.loc[0, 'genres']
```

```python
new_df["crew"] = new_df["crew"].apply(get_names)
```

```python
new_df.loc[0, 'crew']
```

```python
new_df["production_companies"] = new_df["production_companies"].apply(get_names)
```

```python
new_df.loc[0, 'production_companies']
```

```python
new_df.head(1)
```

```python
new_df['overview'] = new_df['overview'].apply(lambda x:str(x).replace(",", "").replace(" ",",").lower())
```

```python
new_df["tags"] = new_df["genres"] + "," + new_df["keywords"] + "," + new_df["overview"] + "," + new_df["cast"] + "," + new_df["crew"] + "," + new_df["production_companies"]
```

```python
final_df = new_df[["id", "title", "tags"]]
```

```python
final_df.dropna(inplace = True)
```

```python
final_df.drop_duplicates(inplace = True)
```

```python
final_df.info()
```

```python
for i in final_df.index:
    print(final_df.loc[i, "title"], "-", final_df.loc[i, "tags"].count(","))
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
# Download stopwords if not already available
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
def remove_stop_words(text):
    # Get the English stopwords list
    stop_words = set(stopwords.words('english'))
    # Tokenize the sentence
    words = text.split(',')
    # Remove stopwords
    filtered_words = [word for word in words if word.lower() not in stop_words]
    # # Convert back to string
    filtered_text = ",".join(filtered_words)
    return filtered_text
```

```python
final_df["tags"] = final_df["tags"].apply(remove_stop_words)
```

```python
final_df.to_csv("cleaned_data.csv")
```
